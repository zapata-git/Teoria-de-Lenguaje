*Teniendo en cuenta los algoritmos de ordenamiento rápido y ordenamiento por burbuja, así como el algoritmo de multiplicación de dos matrices nxn, implementar barómetros en estos que nos permitan contar el número de veces que se ejecuta una instrucción. El objetivo es determinar el orden de complejidad de los tres algoritmos y posteriormente hacer una comparación de ellos.*




***Importar librerias***

```python
import matplotlib.pyplot as plt
import random
import numpy as np
import time
```

**Q-sort**

```python
def qsort(arr):
    contador = 0
    if len(arr) <= 1:
        return arr, contador
    pivote = arr[0]
    izquierda = [x for x in arr[1:]]
    derecha = [x for x in arr[1:] if x > pivote]
    contador += 2 * (len(arr) - 1)
    izquierda, c1 = qsort(izquierda)
    derecha, c2 = qsort(derecha)
    contador += c1 + c2
    return izquierda + [pivote] + derecha, contador
```

**X-sort**

```python
def xsort(arr):
    contador = 0
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            contador += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        contador += 3
    return arr, contador
```

**Multiplicacion de matrices**

```python
def multiplicacion_matrices(a, b):
    contador = 0
    result = np.zeros((a.shape[0], b.shape[1]))
    for i in range(a.shape[0]):
        for j in range(b.shape[1]):
            for k in range(a.shape[1]):
                result[i][j] += a[i][k] * b[k][j]
                contador += 1
    return result, contador
```

**Graficar**

```python
def graficar_complejidad(titulo, color, x, y, label):
    fig, ax = plt.subplots()
    ax.plot(x, y, '-o', color=color, markersize=5, label=label)
    ax.set_xlabel("Numero de datos")
    ax.set_ylabel("Numero de operaciones")
    ax.set_title(titulo)
    ax.legend()
    plt.grid(True)
    plt.show()

x = []
y_qsort = []
y_xsort = []
y_multiplicacion = []
for i in range(1, 101):
    datos = random.sample(range(1, 10001), i)
    a = np.random.rand(i, i)
    b = np.random.rand(i, i)
    _, c = qsort(datos)
    y_qsort.append(c)
    _, c = xsort(datos)
    y_xsort.append(c)
    _, c = multiplicacion_matrices(a, b)
    y_multiplicacion.append(c)
    x.append(i)

graficar_complejidad("Qsort", "red", x, y_qsort, "Qsort")
graficar_complejidad("Xsort", "blue", x, y_xsort, "Xsort")
graficar_complejidad("Multiplicación de matrices", "green", x, y_multiplicacion, "Multiplicación de matrices")
```
